```python
import numpy as np
import pandas as pd
import cv2 
import argparse
import matplotlib.pyplot as plt
import os 
```

Masking `benign` files

```python

dir = '/Users/rhidita/Desktop/BC_FL/Dataset_BUSI_with_GT/benign/'
save_dir_benign = '/Users/rhidita/Desktop/BC_FL/Dataset_Masked/benign'
for i in range(1,438):
    os.chdir(dir)
    image_name = dir+"/benign ({}).png".format(i)
    image_mask = dir+'/benign ({})_mask.png'.format(i)
    filename = 'benign_{}.png'.format(i)
    image = cv2.imread(image_name)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_maskk = cv2.imread(image_mask)

    #checking whether a second and third mask exists or not
    if os.path.exists(dir+'/benign ({})_mask_1.png'.format(i)):
        secondary_mask = cv2.imread(dir+'/benign ({})_mask_1.png'.format(i))
        image_maskk = cv2.bitwise_or(image_maskk,secondary_mask)
        #checking for a 3rd mask:
        if os.path.exists(dir+'/benign ({})_mask_2.png'.format(i)):
            tertiary_mask = cv2.imread(dir+'/benign ({})_mask_2.png'.format(i))
            image_maskk = cv2.bitwise_or(image_maskk,tertiary_mask)



    image_maskk = cv2.cvtColor(image_maskk,cv2.COLOR_BGR2RGB)
    masked_image = cv2.bitwise_and(image,image_maskk)

    #checking whether a second mask exists or not

    os.chdir(save_dir_benign)
   
    cv2.imwrite(filename,masked_image)

```
